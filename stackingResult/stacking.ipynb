```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
train_set_original_path = '/home/adeliaherlisa/Documents/s2/2018 genap/machine learning/TK Akhir/datasets/balancedData/balancedValidation.csv'
train_original_data = pd.read_csv(train_set_original_path, sep=',')
train_original_data.head()
```

```python
path = '/home/adeliaherlisa/Documents/s2/2018 genap/machine learning/TK Akhir/result/stacking/'

train_set_path = path+'pred-train-stacking-dtc45-all.csv'
train_data = pd.read_csv(train_set_path, sep=',', header=None)
train_data.head()
```

```python
test_set_original_path = '/home/adeliaherlisa/Documents/s2/2018 genap/machine learning/TK Akhir/datasets/balancedData/balancedDataTest.csv'
test_original_data = pd.read_csv(test_set_original_path, sep=',')
test_original_data.head()
```

```python
test_set_path = path+'pred-test-stacking-dtc45-all.csv'
test_data = pd.read_csv(test_set_path, sep=',', header=None)
test_data.head()
```

```python
# init metaclassifier

metaLogReg = LogisticRegression().fit(train_data, train_original_data['y'])
metaLinReg = LinearRegression().fit(train_data, train_original_data['y'])
metaMNB = MultinomialNB().fit(train_data, train_original_data['y'])
metaKNN = KNeighborsClassifier().fit(train_data, train_original_data['y'])
metaSVM = svm.SVC(gamma='scale', kernel='rbf').fit(train_data, train_original_data['y'])
metaMLP = MLPClassifier(activation='relu', solver='adam', alpha=1e-5, hidden_layer_sizes=(5, 5), random_state=1).fit(train_data, train_original_data['y'])
```

```python
# predict testing data

predictionLogReg = metaLogReg.predict(test_data)
print(str(round(accuracy_score(test_original_data['y'], predictionLogReg)*100, 2))
      + "\t" + str(round(precision_score(test_original_data['y'], predictionLogReg)*100, 2)) 
      + "\t" + str(round(recall_score(test_original_data['y'], predictionLogReg)*100, 2))
      + "\t" + str(round(f1_score(test_original_data['y'], predictionLogReg)*100, 2))
     )

predictionLinReg = metaLinReg.predict(test_data)
print(str(round(accuracy_score(test_original_data['y'], predictionLinReg.round())*100, 2))
      + "\t" + str(round(precision_score(test_original_data['y'], predictionLinReg.round())*100, 2)) 
      + "\t" + str(round(recall_score(test_original_data['y'], predictionLinReg.round())*100, 2))
      + "\t" + str(round(f1_score(test_original_data['y'], predictionLinReg.round())*100, 2))
     )

predictionMLP = metaMLP.predict(test_data)
print(str(round(accuracy_score(test_original_data['y'], predictionMLP)*100, 2))
      + "\t" + str(round(precision_score(test_original_data['y'], predictionMLP)*100, 2)) 
      + "\t" + str(round(recall_score(test_original_data['y'], predictionMLP)*100, 2))
      + "\t" + str(round(f1_score(test_original_data['y'], predictionMLP)*100, 2))
     )

predictionMNB = metaMNB.predict(test_data)
print(str(round(accuracy_score(test_original_data['y'], predictionMNB)*100, 2))
      + "\t" + str(round(precision_score(test_original_data['y'], predictionMNB)*100, 2)) 
      + "\t" + str(round(recall_score(test_original_data['y'], predictionMNB)*100, 2))
      + "\t" + str(round(f1_score(test_original_data['y'], predictionMNB)*100, 2))
     )

predictionKNN = metaKNN.predict(test_data)
print(str(round(accuracy_score(test_original_data['y'], predictionKNN)*100, 2))
      + "\t" + str(round(precision_score(test_original_data['y'], predictionKNN)*100, 2)) 
      + "\t" + str(round(recall_score(test_original_data['y'], predictionKNN)*100, 2))
      + "\t" + str(round(f1_score(test_original_data['y'], predictionKNN)*100, 2))
     )

predictionSVM = metaSVM.predict(test_data)
print(str(round(accuracy_score(test_original_data['y'], predictionSVM)*100, 2))
      + "\t" + str(round(precision_score(test_original_data['y'], predictionSVM)*100, 2)) 
      + "\t" + str(round(recall_score(test_original_data['y'], predictionSVM)*100, 2))
      + "\t" + str(round(f1_score(test_original_data['y'], predictionSVM)*100, 2))
     )
```
